```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xlrd
```

```python
dat = pd.read_csv("../data/jj_21.csv")
dat["Suspect/interacting Drug List (Drug Char - Indication PT - Action taken - [Duration - Dose - Route])"]= "J&J"

dat1 = pd.read_csv("../data/pfizer_21_1.csv")
dat1_2 = pd.read_csv("../data/pfizer_21_2.csv")

dat2 = pd.read_csv("../data/astra_21_1.csv")
dat2_2 = pd.read_csv("../data/astra_21_2.csv")

dat3 = pd.read_csv("../data/moderna_21.csv")
dat3["Suspect/interacting Drug List (Drug Char - Indication PT - Action taken - [Duration - Dose - Route])"]= "MODERNA"

dat= pd.concat([dat,dat1,dat1_2,dat2,dat2_2,dat3])
dat["EV Gateway Receipt Date"] = pd.to_datetime(dat["EV Gateway Receipt Date"],format='%Y-%m-%d')
#dat=dat.set_index("EV Gateway Receipt Date")
dat["num"]=1
dat
```

```python
test=dat[dat['Reaction List PT (Duration – Outcome - Seriousness Criteria)'].str.contains("hreatening|pitaliza|Disabling|Fatal|Death",na=False)]

test_sum=test.groupby(["EV Gateway Receipt Date","Patient Age Group",'Reaction List PT (Duration – Outcome - Seriousness Criteria)',"Suspect/interacting Drug List (Drug Char - Indication PT - Action taken - [Duration - Dose - Route])"]).agg("sum").reset_index()
test_sum
```

```python
age_dat=dat.groupby(["EV Gateway Receipt Date","Patient Age Group"]).agg("sum").reset_index().set_index("EV Gateway Receipt Date")
age_dat
age_groups=age_dat["Patient Age Group"].unique()

candidates=age_groups[[4,5,6,7]]
age_groups[[4,5,6,7]]
```

```python
st="|".join(candidates)
test_sum2=test_sum[test_sum["Patient Age Group"].str.contains(st,na=False)]
test_sum2
```

```python
adverse =[]
for i in range(len(test_sum2)):
    te=test_sum2.iloc[i,2].split("<BR><BR>")
    #print(len(te))
    for l in range(len(te)):
        adverse.append(te[l].split(" (")[0])
```

```python
un_symptoms=pd.Series(adverse).unique()
sym =[]
n =[]

for i in range(len(un_symptoms)):
    n.append(len(pd.Series(adverse)[pd.Series(adverse)==un_symptoms[i]]))
    sym.append(un_symptoms[i])

adverse_num_df = pd.DataFrame({"symptom":pd.Series(sym),"n":pd.Series(n)})
adverse_num_df=adverse_num_df.sort_values(by=['n'])
adverse_num_df
```

```python
test_sum3=test_sum2.groupby(["EV Gateway Receipt Date","Patient Age Group","Suspect/interacting Drug List (Drug Char - Indication PT - Action taken - [Duration - Dose - Route])"]).agg("sum").reset_index().set_index("EV Gateway Receipt Date")##############

test_sum3["abs"]=0
test_sum3
```

```python
for i in range(len(test_sum3)):
    if i == 0:
        test_sum3.iloc[i,3] = test_sum3.iloc[i,2]
    else: 
        test_sum3.iloc[i,3] = test_sum3.iloc[i,2]+test_sum3.iloc[i-1,3]

test_sum3
```

```python
suppe=test_sum3["Suspect/interacting Drug List (Drug Char - Indication PT - Action taken - [Duration - Dose - Route])"].unique()#######
suppe
#astra komplikationen
ast=test_sum3[test_sum3["Suspect/interacting Drug List (Drug Char - Indication PT - Action taken - [Duration - Dose - Route])"]==suppe[0]].num.sum()##################
#pfizer komplikationen
pf=test_sum3[test_sum3["Suspect/interacting Drug List (Drug Char - Indication PT - Action taken - [Duration - Dose - Route])"]==suppe[1]].num.sum()##################
#astra komplikationen
mod=test_sum3[test_sum3["Suspect/interacting Drug List (Drug Char - Indication PT - Action taken - [Duration - Dose - Route])"]==suppe[2]].num.sum()##################



```

```python
# from matplotlib import gridspec

# from matplotlib.backends.backend_pdf import PdfPages
# # The PDF document
# pdf_pages = PdfPages('EMA_kids_Auswertungen.pdf')
```

```python
fig, (ax,ax1) = plt.subplots(2,1,figsize=(16,9),sharex=True)

plt.suptitle("schwere Nebenwirkungen in den Altersgruppen [0-17] kumulativ und pro Datum ",fontsize=20)

txt = "Alle Fälle in den Altersgruppen 0-17, deren Diagnose irgendeinen der folgende Wörter enthalten: \"Death|Fatal|Life Threatening|Hospitalization|Disabling\"\nQuelle: EMA-Datenbank www.adrreports.eu/"


ax.set_title("kumulativ")
ax.plot(test_sum3.index,test_sum3["abs"],label="kumulativ")

import matplotlib.dates as mdates
# set monthly locator
ax.xaxis.set_major_locator(mdates.DayLocator(interval=7))
# set formatter
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%B-%Y'))
# set font and rotation for date tick labels
plt.gcf().autofmt_xdate()

ax1.set_title("pro Datum und Altersgruppe")
for i,group in enumerate(candidates):
    tempo=test_sum3[test_sum3["Patient Age Group"] == group]
    ax1.bar(tempo.index,tempo["num"],label=group)



plt.figtext(0.1, 0.01, txt, wrap=True, horizontalalignment='left', fontsize=10)

ax.set_ylabel('Anzahl')
ax1.set_ylabel('Anzahl')
plt.legend(loc="best")
fig.show()

fig.savefig("img/kids_overview.svg",format="svg", dpi=1200)
```

```python

# Data to plot
labels = suppe
sizes = [int(ast), int(pf), int(mod)]
colors = ['gold', 'yellowgreen', 'lightcoral']
explode = (0.1, 0.1, 0.1)  # explode 1st slice

# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n(n= {:d} )".format(pct, absolute)

# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "black" }
#plt.rcParams['font.size'] = 20.0
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(sizes, 
                                  autopct = lambda pct: func(pct, sizes),
                                  explode = explode, 
                                  labels = suppe,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="black",size=15))
  
# Adding legend
ax.legend(wedges, suppe,
          title ="Suppen",
          loc ="upper left",
          bbox_to_anchor =(1, 0, 0.5, 1))
  
plt.setp(autotexts, size = 12, weight ="bold")
ax.set_title("Welche Suppe hat die meisten schweren Nebenwirkungen bei Kindern und Jugendlichen?")
txt = "Alle Fälle in den Altersgruppen 0-17, deren Diagnose irgendeinen der folgende Wörter enthalten: \"Death|Fatal|Life Threatening|Hospitalization|Disabling\"\nQuelle: EMA-Datenbank www.adrreports.eu/"

plt.figtext(0.1, 0.01, txt, wrap=True, horizontalalignment='left', fontsize=12)
# show plot
plt.show()
#pdf_pages.savefig(fig)
fig.savefig("img/kids_pie.svg",format="svg", dpi=1200)
```
